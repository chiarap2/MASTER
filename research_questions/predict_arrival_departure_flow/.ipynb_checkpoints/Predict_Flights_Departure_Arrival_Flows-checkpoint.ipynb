```python
import os
import pandas as pd

import datetime
import matplotlib.pyplot as plt

import numpy as np
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
```

### Import the dataset
So far the dataset has aroud 3 years: 2017, 2018 and 2019.1. <br/>
The dataset has been collected from here: http://www.ypa.gr/en/profile/statistics/yearstatistics/

```python
data_path = '/home/master/santorini_analysis/datasets/input/flights_arrivals_and_departures'
os.listdir(data_path)
```

```python
def date_parser(date_time_str):
    a, b = date_time_str.split('-')
    date_time_str = a + '-20' + b
    return datetime.datetime.strptime(date_time_str, '%b-%Y')

lst_dfs = []
for data_file in [d for d in os.listdir(data_path) if ".csv" in d]:
    # print(data_file)
    df = pd.read_csv(os.path.join(data_path, data_file), sep=';', parse_dates=[7], date_parser=date_parser, index_col=7)
    lst_dfs.append(df)
    
df_data = pd.concat(lst_dfs)
df_data.sort_index(inplace=True)
df_data.head(5)
```

```python
df_data.columns
```

### TimeSeries

```python
# Use seaborn style defaults and set the default figure size
sns.set(rc={'figure.figsize':(11, 4)})
import plotly.express as px

df_data['Date'] = df_data.index
```

```python
fig = px.line(df_data, x="Date", y="domestic_flights_arr_dep", color='Zone',width=1000, height=600).update_traces(mode='lines+markers')
fig.update_layout(title_text='Domestic_Flights',
                  xaxis_rangeslider_visible=True)

fig.show()
```

```python
# auxiliar function
def plot_multiple_timeseries(df_ts, ylabel):

    start = min(df_ts.index)
    end = max(df_ts.index)

    fig, ax = plt.subplots()
    for col in list(df_ts.columns):
        ax.plot(df_ts.loc[start:end, col],
        marker='.', linestyle='-', linewidth=0.5, label=col)

    ax.set_ylabel(ylabel)
    ax.legend();

    fig.set_size_inches(18.5, 10.5, forward=True)
    
def get_timeserie_dataframe(df, cat_column, value_column):
    lst_cat = df_data[cat_column].unique()
    lst_dfs = []

    for elem_cat in lst_cat:
        df = df_data[df_data[cat_column] == elem_cat][value_column]
        lst_dfs.append(df)

    df = pd.concat(lst_dfs, axis=1).reindex(lst_dfs[0].index)
    df.columns = lst_cat
    return df
```
