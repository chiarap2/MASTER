# Frequent Pattern mining
For each user geolocated (in Santorini or not) with more than one tweet we compute the sequence of visits

```python
sequence={}
for el in gdf_tweets_geo_tagged[['user_id','name']].dropna().values:
    if not(el[0] in sequence):
        sequence[el[0]]=[]
    sequence[el[0]].append(el[1])

len(sequence)
```

```python
#http://www.borgelt.net/pyfim.html other implementation
import pandas as  pd
import pyfpgrowth
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from collections import Counter

f=2#threshold
dataset = [sequence[el] for el in sequence.keys() if len(sequence[el])>f]

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df.index=[el for el in sequence.keys() if len(sequence[el])>f]
min_support=0.01
frequent_itemsets = apriori(df, min_support=min_support, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
print ("frequent itemset at min support = %s" %(min_support))
print(frequent_itemsets)
df['estimated_country_of_residence']=[country_of_res_dict[x]['Estimated Country of Residence'] if x in country_of_res_dict else None for x in list(df.index)]
```

```python
## Statistics on results
i=32 #pattern id
print ('number of users/sequences:', len(df))
df.groupby(list(frequent_itemsets['itemsets'].ix[i])).count().head()
```

```python
print ('number of users/sequences per country of residence')
I=df[df[list(frequent_itemsets['itemsets'].ix[i])].all(axis=1)][list(frequent_itemsets['itemsets'].ix[i])+['estimated_country_of_residence']]
pd.DataFrame(I.groupby(['estimated_country_of_residence'])['estimated_country_of_residence'].count()).sort_values(by='estimated_country_of_residence',ascending=False).head()
```

```python
def isSantorini(lon,lat):
    if (lon>25.27 and lon<25.52) and (lat>36.30 and lat<36.50):
        return True
    else:
        return False

import datetime
gdf_tweets_geo_tagged['created_at']=gdf_tweets_geo_tagged['created_at'].apply(lambda x:datetime.datetime.strptime(x,'%a %b %d %H:%M:%S +0000 %Y'))
gdf_tweets_geo_tagged['Santorini']=gdf_tweets_geo_tagged['geo_coordinates'].apply(lambda x: isSantorini(x[1],x[0]))
print ("tweets from Santorini", len(gdf_tweets_geo_tagged[gdf_tweets_geo_tagged['Santorini']]), 'over ',len(gdf_tweets_geo_tagged))

#join with country of residence
gdf_tweets_geo_tagged['estimated_country_of_residence']=gdf_tweets_geo_tagged['user_id'].apply(lambda x: country_of_res_dict[x]['Estimated Country of Residence'] if x in country_of_res_dict else None)
```

```python
#change projection and filter tweets from santorini
to_cluster=gdf_tweets_geo_tagged[gdf_tweets_geo_tagged['Santorini']].to_crs(epsg=23034)[['id','geometry']]

#clustering
from sklearn.cluster import DBSCAN
import numpy as np
X = np.array([[el.x,el.y] for el in to_cluster['geometry'].values])
clustering = DBSCAN(eps=150, min_samples=3).fit(X)
to_cluster['labels']=clustering.labels_

#collecting results
to_cluster.index=to_cluster.id
to_cluster=to_cluster.to_dict(orient='index')
gdf_tweets_geo_tagged['cluster_id']=gdf_tweets_geo_tagged['id'].apply(lambda x: to_cluster[x]['labels'] if x in to_cluster else None )

#statistics
cluster_stats=pd.DataFrame(gdf_tweets_geo_tagged[gdf_tweets_geo_tagged['Santorini']].groupby(['cluster_id'])['user_id'].count())
cluster_stats=cluster_stats.sort_values(by='user_id',ascending=False)
cluster_stats.columns=['size']
cluster_stats=cluster_stats.to_dict(orient='index')

#exporting to Qgis
#gdf_tweets_geo_tagged[gdf_tweets_geo_tagged['Santorini']][['user_screen_name','user_id','geometry','created_at','lang','place_place_type','estimated_country_of_residence','cluster_id']].to_file('../datasets/shp/tweets_santorini.shp')

#compute envelopes
cluster_stats_geo=gdf_tweets_geo_tagged[['cluster_id','geometry','user_id']].dissolve(by='cluster_id',aggfunc='count').reset_index()
cluster_stats_geo['geometry']=cluster_stats_geo['geometry'].apply(lambda x: x.envelope)
cluster_stats_geo[cluster_stats_geo['cluster_id']!=-1].plot()
```
